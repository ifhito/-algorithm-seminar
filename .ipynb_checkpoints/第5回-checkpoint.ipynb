```python
# stack
class StackNode:
    def __init__(self, x, y = None):
        self.value = x
        self.next = y
        
class Stack:
    def __init__(self):
        self.top = None
    
    def pop():
        if self.top == None:
            return None
        val = self.top.value
        self.top = self.top.next
        return val
    
    def push(x):
        self.top = StackNode(x, self.top)
    
    def peek():
        if self.top == None:
            return None
        return self.top.value
    
    def isEmpty():
        return self.top == None   
```

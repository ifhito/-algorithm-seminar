```python
#3.4
class MyQuery:
    def __init__(self):
        self.first = Stack()
        self.second = Stack()
        self.flag = False
        
    def enqueue(self, item):
        if self.flag == True:
            while True:
                pop_num = self.second.pop()
                if pop_num == None:
                    break
                self.first.push(pop_num)
            self.flag = False
        newNode = StackNode(item)
        self.first.push(newNode.value)
            
        
    def dequeue(self):
        if self.first.top == None and self.second.top == None:
            return None
        if self.flag == True:
            return self.second.pop()
        else:
            while True:
                pop_num = self.first.pop()
                if pop_num == None:
                    break
                self.second.push(pop_num)
            self.flag = True
            return self.second.pop()
    
    def remove(self):
        if self.first.top == None and self.second.top == None:
            return None
        if self.flag == True:
            while True:
                pop_num = self.second.pop()
                if pop_num == None:
                    break
                self.first.push(pop_num)
            self.flag = False
        data = self.first.pop()
        return data
    
    def peek(self):
        if self.first.top == None and self.second.top == None:
            return None
        if self.flag == True:
            while True:
                pop_num = self.second.pop()
                if pop_num == None:
                    break
                self.first.push(pop_num)
            self.flag = False
        return self.first.top.value
    
    def isEmpty(self):
        return self.first.top == None and self.second.top == None
```

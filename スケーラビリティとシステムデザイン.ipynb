# 9.5

### processSearch(string query)で検索クエリを受付、結果を返すのに100台のマシンを使用
### キャッシュのシステム設計
### キャッシュの更新の仕方については必ず説明

サイトの情報をフロント側で保持しておく→保持するときはDOM情報の要素を連結リストで保存する。また、保存の際、ハッシュ値を求めておき、それぞれの要素毎に保存しておく。情報の変化しているかどうかはハッシュの値を再度ページにアクセスした際にそれぞれの要素に対して求め、その値が同じかどうかで判断する。違っていた場合は、DOM要素を連結リストの頭から読み込む際に比較し、違う部分を更新する。

# 9.7

個人の資産管理システム

システムは銀行口座に接続し、諸費傾向を分析、アドバイス

銀行口座との認証には二段階認証を用いる
基本的にネットバンキングを経由してデータを持ってくることになる。ユーザにはアカウントIDとパスワードを聞き、それを用いてネットバンキング内に入り込むこととする。
アカウントIDとパスワードはハッシュ化された辞書として保存しておき、ハッシュ値によって認証を行うようにする。

データの読み込みはユーザが必要な時に行う。同期して行うと書き込み負荷が大きくなる気がする。取得データは以前の更新があるより前のもの以外は取得しない。また、既定の年数のデータは保存しないようにする(データベースのデータ量の削減)
